### List Comprehension
[ expression for item in list if conditional ]

```python
print(my_list)

for num in my_list:
    #print even numbers
    if num%2 == 0:
        print(num)
```

```python
new_list = [num for num in my_list if num%2==0]
print(new_list)
```

```python
squares_list = [num**2 for num in my_list]
print(squares_list)
```

### Create Functions

```python
def sum_of_list(a_list):
    sum = 0
    for num in a_list:
        sum = sum + num
        
    return sum
```

```python
print(my_list)
```

```python
sum_list = sum_of_list(my_list)
```

```python
print(sum_list)
```

```python
def list_raised_to(a_list, exponent):
    temp_list = []
    for item in a_list:
        temp_list.append(item**exponent)
        
    return temp_list
```

```python
print(my_list)
power_of_3 = list_raised_to(my_list, 3)
```

```python
print(power_of_3)
```
