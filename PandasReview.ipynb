# Intro to Pandas

```python
import numpy as np
import pandas as pd
```

```python
## Pandas Series
## Series is a one-dimensional array like structure
s = pd.Series(np.random.randn(5))

print("Series of Randon Numbers")
print(s)
print("")

new_dict = {'b' : 1, 'a' : 0, 'c' : 2}
s = pd.Series(new_dict)

print("Series from a Dictionary")
print(s)
```

```python
## Pandas Dataframe
## DataFrame is a two-dimensional array
new_dict = {'one' : [1., 2., 3., 4.],
            'two' : [4., 3., 2., 1.],
            'three' : [7., 8., 9., 10.]}

df = pd.DataFrame(new_dict)

print(df)
```

```python
## Create a new column
df["four"] = 0.0
print(df)
```

```python
## Sum of Rows into another column
df['sum'] = df['one'] + df['two'] + df['three'] + df['four']
print(df)
```

### Cleaning DataFrame

```python
filename = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head()
```

```python
df.dtypes
```

```python
#Dealing with the missing data. Replacing ? in data with NaN
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
## Find and Count Missing Data in Dataframe
## "True" stands for missing value, while "False" stands for not missing value.
missing_data = df.isnull()
missing_data.head(5)
```

```python
missing_data["normalized-losses"].value_counts()
```

```python
## Get list of all columns names
missing_data.columns.values.tolist()
```

```python
## Count the missing values in all columns
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

### How to Deal with missing Data
#### Option 1: Drop
#### Option 2: Replace

```python
## Best option is to replace

## Get Average
avg = df["normalized-losses"].astype("float").mean(axis =0)
print("Average", avg)

## Replace NaN in "normalized-losses" with average
df["normalized-losses"].replace(np.nan, avg, inplace = True)
df.head()
```

### Correct data format

```python
#Convert the objects types that are suppose to be ints or floats
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
